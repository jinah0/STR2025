```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 파일 로드 (Colab 환경에 파일을 업로드해야 합니다)
df = pd.read_csv('final_train.csv', encoding='euc-kr')

# 2. 데이터 전처리 및 타입 확인
# 날짜/시간 관련 변수는 필요에 따라 처리합니다.
df['DATE'] = pd.to_datetime(df['DATE'])

# RIDE_DEMAND를 제외한 수치형 변수만 선택
numeric_cols = df.select_dtypes(include=np.number).columns
features_for_corr = numeric_cols.drop(['RIDE_DEMAND', 'ALIGHT_DEMAND']) # 종속변수와 하차인원 제외

# 3. 상관 분석
correlation_matrix = df[features_for_corr.tolist() + ['RIDE_DEMAND']].corr()
ride_demand_corr = correlation_matrix['RIDE_DEMAND'].sort_values(ascending=False)

print("--- RIDE_DEMAND와의 상관계수 순위 ---")
print(ride_demand_corr.drop('RIDE_DEMAND'))

# 4. 상관관계가 높은 상위 특징값 시각화 (태블로 대신 Python으로 간략히)
# 절대값 기준 상위 5개 특징값 선택
top_5_features = ride_demand_corr.drop('RIDE_DEMAND').abs().nlargest(5).index.tolist()

plt.figure(figsize=(15, 10))
plt.suptitle('RIDE_DEMAND와 주요 특징값 간의 산점도', fontsize=16)

for i, col in enumerate(top_5_features):
    plt.subplot(2, 3, i + 1)
    sns.scatterplot(x=df[col], y=df['RIDE_DEMAND'], alpha=0.5)
    plt.title(f'RIDE_DEMAND vs {col} (Corr: {ride_demand_corr[col]:.3f})')
    plt.xlabel(col)
    plt.ylabel('RIDE_DEMAND')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 1. 특징값 선택 및 데이터 준비
# 'TIME' (시간대), 'pyungil' (평일), 'Avg_temp' (평균 기온)과
# 'RIDE_DEMAND_WEEK_PRIOR' (이전 주 승차 인원) 등을 주요 특징으로 사용

features = ['TIME', 'pyungil', 'Avg_temp', 'RIDE_DEMAND_WEEK_PRIOR']
X = df[features]
y = df['RIDE_DEMAND']

# 결측치가 있다면 제거 (회귀 분석 전 필수)
data = pd.concat([X, y], axis=1).dropna()
X = data[features]
y = data['RIDE_DEMAND']

# 2. 통계적 회귀 모델 구축 (Statsmodels 사용)
# Statsmodels는 통계적 검정치(p-value, R-squared)를 쉽게 얻을 수 있어 해석에 유리합니다.
X_with_const = sm.add_constant(X) # 절편(constant)을 추가

# OLS(Ordinary Least Squares, 최소제곱법) 모델 적합
model = sm.OLS(y, X_with_const).fit()

print("\n--- 다중 선형 회귀 분석 결과 (RIDE_DEMAND 예측) ---")
print(model.summary())

# 3. 해석을 위한 모델 예측 및 평가 (Scikit-learn 사용)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

# 평가 지표
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"\n- Test Set MSE: {mse:.2f}")
print(f"- Test Set R-squared: {r2:.4f}")
print(f"- Coefficients: {dict(zip(features, lr.coef_))}")
```
